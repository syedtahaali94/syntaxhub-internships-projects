```python
# Data Smoothing using Binning - Temperature Data

import numpy as np

# Original dataset - temperatures from day 1 to 8
data = [22, 24, 25, 28, 30, 32, 35, 40]

# Sorting the data (though temperature data is already in order by day)
data.sort()

print("Original Temperature Data (by day):")
for day, temp in enumerate(data, 1):
    print(f"Day {day}: {temp}°C")

# Number of bins - let's create 3 bins since we have 8 data points
bin_size = 3  # This will create bins of size 3, 3, and 2

# Create bins
bins = [data[i:i + bin_size] for i in range(0, len(data), bin_size)]

print("\n" + "="*50)
print("DATA SMOOTHING USING BINNING")
print("="*50)

print(f"\nBins (bin size = {bin_size}):")
for i, bin_data in enumerate(bins, 1):
    print(f"Bin {i}: {bin_data}")

# Data smoothing using binning
print("\n" + "-"*50)
print("SMOOTHING RESULTS")
print("-"*50)

# Method 1: Smoothing by bin mean
print("\n1. Smoothing by Bin Mean:")
smoothed_mean = []
bin_means = []
for i, bin_data in enumerate(bins, 1):
    mean_value = np.mean(bin_data)
    bin_means.append(mean_value)
    smoothed_mean.extend([round(mean_value, 2)] * len(bin_data))
    print(f"Bin {i} mean: {round(mean_value, 2)}°C")

# Method 2: Smoothing by bin median
print("\n2. Smoothing by Bin Median:")
smoothed_median = []
bin_medians = []
for i, bin_data in enumerate(bins, 1):
    median_value = np.median(bin_data)
    bin_medians.append(median_value)
    smoothed_median.extend([median_value] * len(bin_data))
    print(f"Bin {i} median: {median_value}°C")

# Method 3: Smoothing by bin boundaries
print("\n3. Smoothing by Bin Boundaries:")
smoothed_boundaries = []
for i, bin_data in enumerate(bins, 1):
    min_val = min(bin_data)
    max_val = max(bin_data)
    print(f"Bin {i} boundaries: min={min_val}°C, max={max_val}°C")
    for val in bin_data:
        if abs(val - min_val) < abs(val - max_val):
            smoothed_boundaries.append(min_val)
        else:
            smoothed_boundaries.append(max_val)

# Display comparison in a table format
print("\n" + "="*70)
print("COMPARISON TABLE")
print("="*70)
print(f"{'Day':<6} {'Original':<12} {'Bin Mean':<12} {'Bin Median':<12} {'Bin Boundaries':<15}")
print("-"*70)

for i in range(len(data)):
    day = i + 1
    original = data[i]
    mean_val = smoothed_mean[i]
    median_val = smoothed_median[i]
    boundary_val = smoothed_boundaries[i]

    # Add indicators for changed values
    mean_change = " ✓" if original != mean_val else ""
    median_change = " ✓" if original != median_val else ""
    boundary_change = " ✓" if original != boundary_val else ""

    print(f"{day:<6} {original:<12} {mean_val:<12}{mean_change} {median_val:<12}{median_change} {boundary_val:<15}{boundary_change}")

# Summary
print("\n" + "="*50)
print("SUMMARY")
print("="*50)
print(f"Original temperatures: {data}")
print(f"Smoothed (mean):      {[round(x, 1) for x in smoothed_mean]}")
print(f"Smoothed (median):    {smoothed_median}")
print(f"Smoothed (boundaries): {smoothed_boundaries}")

# Calculate and display smoothing impact
print("\n" + "-"*50)
print("SMOOTHING IMPACT")
print("-"*50)

original_mean = np.mean(data)
original_std = np.std(data)

mean_smoothed_mean = np.mean(smoothed_mean)
mean_smoothed_std = np.std(smoothed_mean)

median_smoothed_mean = np.mean(smoothed_median)
median_smoothed_std = np.std(smoothed_median)

boundary_smoothed_mean = np.mean(smoothed_boundaries)
boundary_smoothed_std = np.std(smoothed_boundaries)

print(f"Original data    - Mean: {original_mean:.2f}°C, Std Dev: {original_std:.2f}")
print(f"Bin mean method  - Mean: {mean_smoothed_mean:.2f}°C, Std Dev: {mean_smoothed_std:.2f} (Reduced variability)")
print(f"Bin median method- Mean: {median_smoothed_mean:.2f}°C, Std Dev: {median_smoothed_std:.2f} (Reduced variability)")
print(f"Bin boundaries   - Mean: {boundary_smoothed_mean:.2f}°C, Std Dev: {boundary_smoothed_std:.2f} (Reduced variability)")
```

```python
# Min-Max Normalization

data = [22, 24, 25, 28, 30, 32, 35, 40]

min_val = min(data)
max_val = max(data)

normalized_data = []

for x in data:
    norm = (x - min_val) / (max_val - min_val)
    normalized_data.append(round(norm, 4))  # Rounded to 4 decimal places for cleaner output

print("Original Data:", data)
print("Normalized Data:", normalized_data)

# Optional: Create a table format for better visualization
print("\nDay | Original | Normalized")
print("-" * 30)
for i, (orig, norm) in enumerate(zip(data, normalized_data), 1):
    print(f"{i:3d} | {orig:8d} | {norm:.4f}")
```

```python
import numpy as np

data = [22, 24, 25, 28, 30, 32, 35, 40]

mean = np.mean(data)
std = np.std(data)

zscore_data = []

for x in data:
    z = float((x - mean) / std)
    zscore_data.append(round(z, 4))  # Rounded to 4 decimal places for cleaner output

print("Original Data:", data)
print(f"Mean: {mean:.2f}")
print(f"Standard Deviation: {std:.2f}")
print("Z-Score Normalized Data:", zscore_data)

# Optional: Create a table format for better visualization
print("\nDay | Original | Z-Score")
print("-" * 35)
for i, (orig, z) in enumerate(zip(data, zscore_data), 1):
    print(f"{i:3d} | {orig:8d} | {z:8.4f}")
```
