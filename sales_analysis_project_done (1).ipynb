```python
# Explotary data analysis
# what are the over all sales trends

df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
df['month_year']=df['order_date'].apply(lambda x: x.strftime('%y-%m'))
```

```python
df.groupby('month_year').sum()

```

```python
grouped_data = df.groupby('month_year')[['sales', 'quantity', 'discount', 'profit', 'shipping_cost']].sum()
display(grouped_data)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(grouped_data.index, grouped_data['sales'])
plt.xlabel('Month-Year')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(grouped_data.index, grouped_data['quantity'])
plt.xlabel('Month-Year')
plt.ylabel('Total Quantity')
plt.title('Monthly Quantity Trend')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#which are the top ten products sales
top_ten_products_sales = df.groupby('product_name')['sales'].sum().nlargest(10)
display(top_ten_products_sales)
```

```python
top_ten_products_sales.sort_values(ascending=False)
```

```python
#which are the most selling product
grouped_data=df.groupby('product_name')['quantity'].sum().nlargest(10)
display(grouped_data)
```

```python
grouped_data = grouped_data.sort_values(ascending=False)
display(grouped_data)
```
